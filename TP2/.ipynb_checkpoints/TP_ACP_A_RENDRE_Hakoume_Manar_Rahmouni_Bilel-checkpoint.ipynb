```python
import numpy as np
np.set_printoptions(threshold=10000,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

## Visualisation du fichier

```python
fileVilles = pd.read_csv('villes.csv', sep=';')
fileVilles
```

## Separer les données

Il faut se separer des villes pour pouvoir standardiser les valeurs plus tard (il faut garder uniquement des valeurs)

```python
X = fileVilles.iloc[:, 1:13]
labels = fileVilles.iloc[:, 0]
X
```

## Standardiser les données

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Données avant standardisation :

```python
X1 = pd.DataFrame(X,columns=X.columns)
X1.plot(kind='density',sharex=True,figsize=(12,5),layout=(10,1))
```

### Données apres standardisation :

```python
pd.DataFrame(X_scaled,columns=X.columns).plot(kind='density',sharex=True,figsize=(12,5),layout=(10,1))
```

## Analyse en composantes principales (PCA)

```python
#Import PCA de sklearn
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(X_scaled)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.cumsum()
```

On remarque que avec uniquement le premier axe on arrive deja a 87% de l'information et on y rajoutant le 2ème axe on atteint 98% d'information

#### Visualisation graphique 

```python
scree = pca.explained_variance_ratio_*100
plt.bar(np.arange(len(scree))+1, scree)
plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="green",marker='o')
plt.xlabel("Number of principal components")
plt.ylabel("Percentage explained variance")
plt.title("Scree plot")
plt.show(block=False)
```

À partir du diagramme d'éboulis(scree plot), nous pouvons lire le pourcentage de la variance dans les données expliquées lorsque nous ajoutons des composants principaux. Ainsi, la première composante principale explique 87 % de la variance de l'ensemble de données. Les 2 premières composantes principales expliquent 98 %.

#### Axe 1 

```python
pc1 = pca.components_[0]
pc1
```

Tout les mois sont en correlation positive avec le premier axe.Les valeurs de cette axe correle presque autour de la meme valeur avec une corelation positive allant de 0.27 a 0.30 cet axe 1 est donc aproximativement pareil.

#### Axe 2

```python
pc2= pca.components_[1]
pc2
```

Les valeurs de cette axe ne correle pas de la meme facon, on remarque que les mois janvier, fevrier, mars ,octobre novembre et decembre sont de correlation positives avec le deuxieme axe mais les autres mois sont de correlations negatives.

#### Nouvelle données avec uniquement les 2 axes

```python
new_pca.explained_variance_ratio_
```

#### Variable sur les composants principaux

```python
procedures = new_pca.components_
num_pc = new_pca.n_features_
pc_list =["Y"+str(i) for i in list(range(1,num_pc+1))]
procedures_df=pd.DataFrame.from_dict(dict(zip(pc_list,procedures)))
procedures_df['Mois'] = fileVilles.columns[1:13].values
procedures_df = procedures_df.set_index('Mois')
procedures_df
```

```python
procedures_df.plot.scatter('Y1', 'Y2')
for x,y in procedures_df.iterrows():
    plt.annotate(x,y)
plt.title('Variables correlations sur les composants principaux')
plt.show()
```

## Fonction qui réalise tout le travail précedent :

Pour les données qui ont 3 axes (et non pas 2)

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def fonction_acp(dataset,labels_col,col_variables,mindata_pourcent) :
    dataset = StandardScaler().fit_transform(dataset)
    pca = PCA()
    components = pca.fit_transform(dataset)
    exp_var =pca.explained_variance_ratio_*100
    cum_sum_eigenvalues = np.cumsum(exp_var)
    
    #Visualisation graphique des axes
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="green",marker='o')
    plt.xlabel("Nombres de composants principaux")
    plt.ylabel("Percentage explained variance")
    plt.title("Scree plot")
    plt.show(block=False)
    
    nb_comp =0
    for i in range(len(cum_sum_eigenvalues)) :
        if cum_sum_eigenvalues[i]>=mindata_pourcent :
            nb_comp=i+1
            break
    pca = PCA(n_components=nb_comp)
    components = pca.fit_transform(dataset)
    
    
    #Variables sur les components principaux
    procedures = pca.components_
    num_pc = pca.n_features_
    pc_list =["Y"+str(i) for i in list(range(1,num_pc+1))]
    procedures_df=pd.DataFrame.from_dict(dict(zip(pc_list,procedures)))
    procedures_df['Variables'] = col_variables.values
    procedures_df = procedures_df.set_index('Variables')
    print(procedures_df)
    
    if nb_comp==2 :
        fig = plt.figure(figsize=(8,6))
        ax= fig.add_subplot(1,1,1)
        ax.grid()
        ax.set_xlabel('Composant principal 1 ('+str(round(exp_var[0],2))+'%)')
        ax.set_ylabel('Composant principal 2 ('+str(round(exp_var[1],2))+'%)')
        ax.set_title("2 Components ACP")
    
        plt.scatter(components[:, 0], components[:, 1])
        for l, x, y in zip(labels_col,components[:, 0], components[:, 1]):
            plt.annotate(l, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')
        plt.show()
        #plot variables correlation avec chaque components pricipal
        procedures_df.plot.scatter('Y1', 'Y2')
        for x,y in procedures_df.iterrows():
            plt.annotate(x,y)
        plt.title('Variables correlations sur les composants principaux')
        plt.show()
        
        
    elif nb_comp==3 :
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111, projection='3d')
        for i in range(len(components)): #plot each point + it's index as text above
            ax.scatter(components[i,0],components[i,1],components[i,2],color='b') 
            ax.text(components[i,0],components[i,1],components[i,2],  '%s' % (str(labels_col[i])), size=8, zorder=1,  
            color='k') 
        ax.set_title('3 Components ACP')
        ax.set_xlabel("Y1 ("+str(round(exp_var[0],2))+"%)")
        ax.set_ylabel("Y2 ("+str(round(exp_var[1],2))+"%)")
        ax.set_zlabel("Y3 ("+str(round(exp_var[2],2))+"%)")
        plt.show()
        #plot variables correlation avec chaque components pricipaux
        fig = plt.figure(figsize=(10,10))
        ax = fig.add_subplot(111, projection='3d')
        for x, y in procedures_df.iterrows():
            ax.scatter(y[0],y[1],y[2],color='b') 
            ax.text(y[0],y[1],y[2], '%s' %(str(x)), size=8, zorder=1,  
            color='k') 
        ax.set_title('Variables corrélations sur les components principaux')
        ax.set_xlabel("Y1 ("+str(round(exp_var[0],2))+"%)")
        ax.set_ylabel("Y2 ("+str(round(exp_var[1],2))+"%)")
        ax.set_zlabel("Y3 ("+str(round(exp_var[2],2))+"%)")
        plt.show()
        
    return procedures_df
```

### Test sur les villes

```python
fonction_acp(X,labels,fileVilles.columns[1:13],90)
```
